```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# 定義關鍵字
keywords = ['社論', '台電', '電價', '水費', '台塑', '塑化', '台化', '南亞', '台塑新智能', '水情', '碳費', 
            '石油行情', '煤價', '寧波', '越南', '風電', '光電', '水電', '儲能', '匯率', '政府內閣人員更新']
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd

# 設置 Chrome 選項
chrome_options = Options()
chrome_options.add_argument("--headless")  # 無頭模式
driverpath = 'D:/chromedriver-win64/chromedriver-win64/chromedriver.exe'
# driverpath = 'C:/chromedriver-win64/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_experimental_option("prefs", {"profile.password_manager_enabled": False, "credentials_enable_service": False})

# 启动 Chrome 浏览器
service = Service(executable_path=driverpath)
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()

# 爬取目標網頁
driver.get("https://www.ctee.com.tw/livenews/policy")

# 等待頁面加載
driver.implicitly_wait(10)  # 最多等待 10 秒

# 查找新聞卡片元素
story_contents = driver.find_elements(By.CLASS_NAME, 'newslist__card')
news_data = []

# 提取標題和連結
for story_content in story_contents:
    title_tag = story_content.find_element(By.CLASS_NAME, 'news-title')
    link = title_tag.find_element(By.TAG_NAME, 'a')
    href = link.get_attribute('href')
    title = link.text.strip()
    # news_data.append({'網站名稱': "工商時報", '標題': title, '連結': href})
    # 檢查標題是否包含關鍵字，並分類
    for keyword in keywords:
        if keyword in title:
            # 將新聞資料加入到暫存列表中
            news_data.append({'網站名稱': "工商時報", '標題': title, '連結': href})
    print("標題:", title)
    print("連結:", href)
    print()

# 將暫存列表轉換成 DataFrame
news_df = pd.DataFrame(news_data)

# 打印整個 DataFrame
news_df

# 關閉 WebDriver
driver.quit()
```
