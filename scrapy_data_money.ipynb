```python
#爬首頁新聞

def clean_text(text):
    return re.sub(r'\s+', '', text)

# 創建集合來追蹤已經爬取過的標題和網址
visited_urls = set()
visited_titles = set()

# 爬取一般新聞
url = 'https://money.udn.com/money/index'
res = requests.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
web = "經濟日報"
news_data = []  # 用來暫時儲存新聞資料的列表
# 找到所有包含故事內容的div
story_contents = soup.find_all('div', class_='story__content')

# 迭代處理每個故事內容
for story_content in story_contents:
    # 找到故事內容中的<a>標籤
    link = story_content.find('a')
    # 如果有找到<a>標籤，則獲取其href和title屬性
    if link:
        href = link['href']
        # 檢查href是否以'https:/'開頭，如果不是則加上
        if not href.startswith('https:/'):
            href = 'https:/' + href
        # 移除換行符號
        href = clean_text(href)
        title = link.get('title', 'No title available')
        # 移除標題中的換行符號
        title = clean_text(title)
        # 檢查標題是否為'No title available'，如果是則跳過
        if title == 'Notitleavailable':
            continue
        # 檢查是否已經爬取過相同的標題或網址，如果是則跳過
        if title in visited_titles or href in visited_urls:
            continue
        # 將標題和網址加入已經爬取過的集合中
        visited_titles.add(title)
        visited_urls.add(href)
        # # 將新聞資料加入到暫存列表中
        # news_data.append({'網站名稱': web, '標題': title, '連結': href})
        # 檢查標題是否包含關鍵字，並分類
        for keyword in keywords:
            if keyword in title:
                # 將新聞資料加入到暫存列表中
                news_data.append({'網站名稱': web, '標題': title, '連結': href})
        # 輸出標題和連結
        print("標題:", title)
        print("連結:", href)
        print()

        
# 將暫存列表轉換成DataFrame
news_df = pd.DataFrame(news_data)

# 打印整個DataFrame
news_df
```

```python
import re
import requests
from bs4 import BeautifulSoup
import pandas as pd

def clean_text(text):
    return re.sub(r'\s+', '', text)

def fetch_news(url, web, container_tag, container_class, link_tag, link_class=None, href_prefix='https:'):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    story_contents = soup.find_all(container_tag, class_=container_class)
    news_data = []
    visited_urls = set()
    visited_titles = set()
    
    for story_content in story_contents:
        link = story_content.find(link_tag, class_=link_class) if link_class else story_content.find(link_tag)
        if link:
            href = link['href']
            if not href.startswith('https:/'):
                href = href_prefix + href
            href = clean_text(href)
            title = link.get('title', 'No title available')
            title = clean_text(title)
            if title == 'No title available' or title in visited_titles or href in visited_urls:
                continue
            visited_titles.add(title)
            visited_urls.add(href)
            for keyword in keywords:
                if keyword in title:
                    news_data.append({'網站名稱': web, '標題': title, '連結': href})
            print("標題:", title)
            print("連結:", href)
            print()
    
    return news_data

# 主程式

web = "經濟日報"
all_news_data = []

# 爬取不同類型的新聞
all_news_data.extend(fetch_news('https://money.udn.com/money/index', web, 'div', 'story__content', 'a'))
all_news_data.extend(fetch_news('https://money.udn.com/money/vip/index', web, 'li', 'vip-story__item', 'a'))
all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5595?from=edn_navibar', web, 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))
all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5588?from=edn_navibar', web, 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))
all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5589?from=edn_navibar', web, 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))

# 將暫存列表轉換成DataFrame
news_df = pd.DataFrame(all_news_data)

# 打印整個DataFrame
news_df
```
