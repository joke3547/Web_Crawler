```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import pandas as pd

# 設置 Chrome 選項
chrome_options = Options()
chrome_options.add_argument("--headless")  # 無頭模式
driverpath = 'D:/chromedriver-win64/chromedriver-win64/chromedriver.exe'
# driverpath = 'C:/chromedriver-win64/chromedriver.exe'
options = webdriver.ChromeOptions()
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)
options.add_experimental_option("prefs", {"profile.password_manager_enabled": False, "credentials_enable_service": False})
```

```python
def clean_text(text):
    return re.sub(r'\s+', '', text)
```

```python
#爬蟲方式 1

def fetch_news(url, container_tag, container_class, link_tag, link_class=None, href_prefix='https:'):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, 'html.parser')
    
    story_contents = soup.find_all(container_tag, class_=container_class)
    news_data = []
    visited_urls = set()
    visited_titles = set()
    
    for story_content in story_contents:
        link = story_content.find(link_tag, class_=link_class) if link_class else story_content.find(link_tag)
        if link:
            href = link['href']
            if not href.startswith('https:/'):
                href = href_prefix + href
            href = clean_text(href)
            title = link.get('title', 'No title available')
            title = clean_text(title)
            if title == 'No title available' or title in visited_titles or href in visited_urls:
                continue
            visited_titles.add(title)
            visited_urls.add(href)
            news_data.append({'網站名稱': "經濟日報", '標題': title, '連結': href})
            # for keyword in keywords:
            #     if keyword in title:
            #         news_data.append({'網站名稱': "經濟日報", '標題': title, '連結': href})
            print("標題:", title)
            print("連結:", href)
            print()
    
    return news_data
```

```python
#爬蟲「經濟日報」

def scrapy_money():

    all_news_data = []
    # 爬取不同類型的新聞
    all_news_data.extend(fetch_news('https://money.udn.com/money/index', 'div', 'story__content', 'a'))
    all_news_data.extend(fetch_news('https://money.udn.com/money/vip/index', 'li', 'vip-story__item', 'a'))
    all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5595?from=edn_navibar', 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))
    all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5588?from=edn_navibar', 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))
    all_news_data.extend(fetch_news('https://money.udn.com/money/cate/5589?from=edn_navibar', 'li', 'story__item story', 'a', href_prefix='https://money.udn.com'))

    # 將暫存列表轉換成DataFrame
    news_money_df = pd.DataFrame(all_news_data)

    return news_money_df
```

```python
#爬蟲方式 2
def setup_driver():
    service = Service(executable_path=driverpath)
    driver = webdriver.Chrome(service=service, options=options)
    driver.maximize_window()
    return driver

def fetch_news(driver, url, web_name, container_class, title_class, link_tag='a', keywords=None):
    driver.get(url)
    driver.implicitly_wait(10)  # 最多等待 10 秒

    visited_urls = set()
    visited_titles = set()
    story_contents = driver.find_elements(By.CLASS_NAME, container_class)
    news_data = []

    for story_content in story_contents:
        title_tag = story_content.find_element(By.CLASS_NAME, title_class)
        link = title_tag.find_element(By.TAG_NAME, link_tag)
        href = link.get_attribute('href')
        title = link.text.strip()
        if title == 'Notitleavailable' or title in visited_titles or href in visited_urls:
            continue
        visited_titles.add(title)
        visited_urls.add(href)
        news_data.append({'網站名稱': web_name, '標題': title, '連結': href})
        # for keyword in keywords:
        #     if keyword in title:
        #         news_data.append({'網站名稱': web_name, '標題': title, '連結': href})
        #         break
        # else:
        #     news_data.append({'網站名稱': web_name, '標題': title, '連結': href})
        print("標題:", title)
        print("連結:", href)
        print()

    return news_data
```

```python
#爬蟲「聯合報」

def scrapy_udn():
    # 設置 Chrome 選項
    chrome_options = Options()
    chrome_options.add_argument("--headless")  # 無頭模式
    driverpath = 'D:/chromedriver-win64/chromedriver-win64/chromedriver.exe'
    # driverpath = 'C:/chromedriver-win64/chromedriver.exe'
    options = webdriver.ChromeOptions()
    options.add_experimental_option("excludeSwitches", ["enable-automation"])
    options.add_experimental_option('useAutomationExtension', False)
    options.add_experimental_option("prefs", {"profile.password_manager_enabled": False, "credentials_enable_service": False})

    # 启动 Chrome 浏览器
    service = Service(executable_path=driverpath)
    driver = webdriver.Chrome(service=service, options=options)
    driver.maximize_window()

    try:
        # 爬取目標網頁
        driver.get("https://udn.com/news/cate/2/6644#sub_7240")

        # 等待頁面加載
        driver.implicitly_wait(10)  # 最多等待 10 秒

        # 創建集合來追蹤已經爬取過的標題和網址
        visited_urls = set()
        visited_titles = set()

        # 查找新聞卡片元素
        story_contents = driver.find_elements(By.CLASS_NAME, 'story-list__news ')
        news_data = []

        # 提取標題和連結
        for story_content in story_contents:
            title_tag = story_content.find_element(By.CLASS_NAME, 'story-list__text')
            link = title_tag.find_element(By.TAG_NAME, 'a')
            href = link.get_attribute('href')
            title = link.text.strip()
            # 檢查標題是否為'No title available'，如果是則跳過
            if title == 'Notitleavailable':
                continue
            # 檢查是否已經爬取過相同的標題或網址，如果是則跳過
            if title in visited_titles or href in visited_urls:
                continue
            # 將標題和網址加入已經爬取過的集合中
            visited_titles.add(title)
            visited_urls.add(href)
            news_data.append({'網站名稱': "聯合報", '標題': title, '連結': href})
            print("標題:", title)
            print("連結:", href)
            print()

        # 將暫存列表轉換成 DataFrame
        news_udn_df = pd.DataFrame(news_data)
        return news_udn_df
        
    finally:
        # 關閉 WebDriver
        driver.quit()
```
